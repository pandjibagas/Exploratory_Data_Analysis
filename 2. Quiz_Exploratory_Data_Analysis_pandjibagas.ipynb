```python
online_bl=online_bl.fillna(value=0)
online_bl[['price_original','price_discount']]=online_bl[['price_original','price_discount']].astype('int')
online_bl['time_update']=pd.to_datetime(online_bl['time_update'])
online_bl['month']=online_bl['time_update'].dt.month_name()
online_bl.pivot_table(index='month',columns=['sub_category','scale'],values='price_original',aggfunc='mean')
```

Read `techcrunch.csv` from `data_input` folder to aswer **question 4-6**, a dataset that stores fundraising rounds and amounts from startup companies of different categories around the US.

```python
## Your code here
tech = pd.read_csv('data_input/techcrunch.csv')
tech['fundedDate']=pd.to_datetime(tech['fundedDate'])
tech['fundedDate_M']=tech['fundedDate'].dt.to_period('M')
tech = tech.fillna(0)
tech.dtypes
tech
```

```python
tech.tail()
```

4. Using `techcrunch.csv`, which `category` raised the most amount in funding (`raisedAmt`) on average (use the `median`)?

    *Berdasarkan data `techcrunch`, kategori (`category`) startup manakah yang mendapatkan rata-rata (gunakan `median`) funding (`raisedAmt`) tertinggi?*
    
    - [ ] `mobile`
    - [ ] `cleantech`
    - [x] `biotech`
    - [ ] `consulting`

```python
tech.pivot_table(index='category',values='raisedAmt',aggfunc='median').sort_values(by='raisedAmt',ascending=False)
```
